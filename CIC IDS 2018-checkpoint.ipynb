```python
def quickML(modelName, model, X, y):
    global pre_processor
    
    X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=0.2, random_state=0)
    
    model = Pipeline(steps=[('preprocessor', pre_processor), (modelName, model)])
    model.fit(X_train, y_train)
    y_pred = model.predict(X_test)

    TN, FP, FN, TP = confusion_matrix(y_test, y_pred, labels=[0, 1]).ravel()
    TPR = TP / (TP + FN)
    TNR = TN / (TN + FP)
    NPV = TN / (TN + FN)
    FPR = FP / (FP + TN)
    NPV = TN / (TN + FN)
    FDR = FP / (TP + FP)
    FNR = FN / (TP + FN)
    specificity = TN / (TN+FP)

    auc = roc_auc_score(y_test, y_pred)   
    cm = confusion_matrix(y_test, y_pred)
    disp = ConfusionMatrixDisplay(cm)    
    disp.plot()
    plt.title('Model: ' + modelName)
    plt.show()
    plt.savefig('results.png')
    print('------ ' + modelName + ' statistical data ------')
    print('Accuracy score: ' + str(accuracy_score(y_test, y_pred)))
    print('Precision score: ' + str(precision_score(y_test, y_pred)))
    print('Recall score: ' + str(recall_score(y_test, y_pred)))
    print('Specificity:' + str(specificity))
    print('f1 score: ' + str(f1_score(y_test, y_pred)))
    print('KAPPA: ' + str(cohen_kappa_score(y_test, y_pred)))
    print('MCC: ' + str(matthews_corrcoef(y_test, y_pred)))
    print('AUC :' + str(auc))
    print('False Discovery Rate: ' + str(FDR))
    print('False Negative Rate: ' + str(FNR))
    print('False Positive Rate: ' + str(FPR))
    print('Negative Predictive Value: ' + str(NPV))
```
